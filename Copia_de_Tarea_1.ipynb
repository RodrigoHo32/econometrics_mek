# RESPUESTA 1

En econometría comenzamos con la revisión de conceptos básicos, para lo cual consideramos este primer modelo, conocido como modelo de regresión constante. Este es un modelo explicativo cuyo objetivo es estimar parámetros poblacionales a partir de una muestra de datos.

Al construir un modelo econométrico, el primer paso consiste en identificar la expresión matemática que describe la relación económica de interés, así como determinar las variables relevantes a incluir. Como menciona Bruce Hansen en su libro ECONOMETRICS, existe un caso especial, que es el modelo de regresión "INTERCEPT-ONLY" el cual no tiene regresosres X, y el valor esperado de la variable dependiente es constante.
$$
Y_i = \alpha + \varepsilon_i
$$

Donde $\alpha=E[Y_i]$ representa la media poblacional y el error debe cumplir con $E[\varepsilon_i]=0$.

En el termino de error se recogen todos los factores o efectos no observados que afectan a $ Y_i $ y que se encuentran fuera del modelo.

Para estimar el parámetro $\alpha$ utilizando el método OLS, de Mínimos Cuadrados Ordinarios, debemos de considerar que el objetivo se centra en minimizar la suma de los residuos al cuadrado entre los valores que estamos observando y el que se está estimando:

$$
\hat{\alpha}=\min_{\alpha}\sum_{i=1}^{n}(Y_i-\alpha)^2
$$


Lo cual se traduce a la función objetivo que se presenta a continuación:
$$
S(\alpha)=\sum_{i=1}^{n}(Y_i-\alpha)^2
$$

Si hacemos la derivación respecto a $\alpha$ e igualando a 0:

$$
\frac{\partial S(\alpha)}{\partial\alpha}
=-2\sum_{i=1}^{n}(Y_i-\alpha)=0
$$

Simplificando el termino, eliminamos el (-2), dividiendo ambos lados de la ecuación entre (-2), y dado que $\sum_{i=1}^{n}(\alpha)$ se puede representaro por $n\alpha$, podemos reducir la expresión de la siguiente manera:
$$
\sum_{i=1}^{n}Y_i-n\alpha=0
$$

Despejando $\alpha$, se obtiene el estimador:
$$
\hat{\alpha}=\frac{1}{n}\sum_{i=1}^{n}Y_i=\bar Y
$$

Por lo anterior, obtenemos:

$$
\hat{\alpha}=\hat{Y}
$$

Este resultado observamos muestra que, en ausencia de regresores, el valor que minimiza los cuadrados de los errores, es la media de las muestras, ya que cualquier otro valor que se estime, tendrá un error más grande, es decir, la suma de los errores al cuadrado, será mayor que la de este escenario.

```python
variables = ["white","black","hispanic","eng_sec_language","havekids_at_bline","educlevel_at_bline","hsgrad_at_bline","ged_at_bline","mos_wrk_lyr_bline",
            "earn_lyr_bline","wage_bline","anywlfr_bline","alcohol_bline","arrest_bline","jail_bline"]

for i in variables:

    ctl = JobCorps.loc[JobCorps.treatmnt == 0, i].dropna()
    tmt = JobCorps.loc[JobCorps.treatmnt == 1, i].dropna()

    p = ttest_ind(ctl, tmt, equal_var=False).pvalue
    t = ttest_ind(ctl, tmt, equal_var=False).statistic

    print(i, "CTL:", ctl.mean(), "TMT:", tmt.mean(), "t_s:", t,"P-value:", p)
```

#### k. Now estimate the experimental impact on earnings in quarters 2-16 after random assignment. What does these result suggest about the time-profile of the effect of being in the Job Corps program? [It may be useful to plot the estimated effect as a function of time]

## Respuesta k

#### Se observa un efecto negativo en los primeros trimestres, que poco a poco y a partir del septimo trimestre, comienzan a ser positivos y aumenta con el tiempo, lo que sugiere una contracción de los ingresos de forma inicial seguido de ganancias en ingresos conforme avanza el programa.

```python
efectos = []

for q in range(1, 17):

    var = f"earnq{q}"
    y = JobCorps_a[var]
    X = sm.add_constant(JobCorps_a["treatmnt"])

    model = sm.OLS(y, X, missing="drop").fit()
    efectos.append(model.params["treatmnt"])

    print(var,"efectos:", round(model.params["treatmnt"],2),"p:", round(model.pvalues["treatmnt"],4))
```

```python
quarters = list(range(1, 17))  # Q1 a Q16

plt.figure()
plt.bar(quarters, efectos)
plt.axhline(0)   # línea en cero

plt.xlabel("Trimestre")
plt.ylabel("Efecto estimado")
plt.title("Ingresos por trimestre")

plt.show()
```
